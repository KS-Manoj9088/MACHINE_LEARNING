```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import  StandardScaler
from sklearn.datasets import load_wine
```

```python
wine = load_wine()
wine_data = pd.DataFrame(wine.data, columns=wine.feature_names)
wine_data.head()
```

```python
#Using dataset read_csv

candy = pd.read_csv("candy-data.csv")
candy.tail()
```

```python
# a. Find missing values
missing_values = candy.isnull().sum()
print(f"Missing values:\n{missing_values}\n")
```

```python
# b. Print the number of rows and columns
print(f"Number of rows: {candy.shape[0]}")
print(f"Number of columns: {candy.shape[1]}\n")
```

```python
# c. Plot box plot for all numerical features
plt.figure(figsize=(12, 6))
sns.boxplot(data=candy.drop(columns=['competitorname']))
plt.title("Box Plot for All Numerical Features")
plt.xticks(rotation=90)
plt.show()
```

```python
# d. Heatmap showing correlations between features
plt.figure(figsize=(12, 8))
sns.heatmap(candy.drop(columns=['competitorname']).corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# e. Scatter plot between two features, e.g., 'chocolate' vs 'caramel'

plt.figure(figsize=(8, 6))
sns.scatterplot(x='chocolate', y='caramel', data=candy)
plt.title("Chocolate vs Caramel")
plt.xlabel('Chocolate')
plt.ylabel('Caramel')
plt.show()
```

```python
# f. Bubble chart: Size of bubbles based on 'sugarpercent'
plt.figure(figsize=(8, 6))
sns.scatterplot(x='chocolate', y='caramel', size='sugarpercent', sizes=(20, 200), data=candy, legend=False)
plt.title("Bubble Chart: Chocolate vs Caramel (Size by Sugar Percent)")
plt.xlabel('Chocolate')
plt.ylabel('Caramel')
plt.show()
```

```python
# g. Area chart for 'sugarpercent' over competitors
plt.figure(figsize=(10, 6))
plt.fill_between(candy['competitorname'], candy['sugarpercent'], color="skyblue", alpha=0.4)
plt.plot(candy['competitorname'], candy['sugarpercent'], color="Slateblue", alpha=0.6, linewidth=2)
plt.title("Area Chart: Sugar Percent Over Competitors")
plt.xlabel('Competitor Name')
plt.ylabel('Sugar Percent')
plt.xticks(rotation=90)
plt.show()
```
