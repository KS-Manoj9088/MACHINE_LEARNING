```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN

data = pd.read_csv('synthetic_dataset.csv')
X = data[['Income', 'Spending Score']]

model_dbscan = DBSCAN(eps=0.1, min_samples=2)
data['DBSCAN Cluster'] = model_dbscan.fit_predict(X)

plt.scatter(data['Income'], data['Spending Score'], c=data['DBSCAN Cluster'], cmap = 'viridis', marker = 'o', s=50)
plt.xlabel('Income')
plt.ylabel('Spending Score')
plt.title('DBSCAN Clustering')
plt.colorbar(label='Cluster Label')
plt.show()
```
